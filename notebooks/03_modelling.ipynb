```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Cleaned_data.csv')
```

```python
print('Data Shape::',data.shape )
```

```python
## we have to encode those categories like BENING to 0 and others to 1...2...etc
print('Labels :',data['Label'].unique())
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
le = LabelEncoder()
data['New_Label']=le.fit_transform(data['Label'])
print(le.classes_)
```

```python
print(f"Total columns: {len(data.columns)}")
print(data.columns.tolist())
```

```python
data_to_dl = data.drop(columns=['Label'])
print('Remaining columns:',len(data_to_dl.columns))
```

```python
X = data_to_dl.drop(columns=['New_Label'])
Y = data_to_dl['New_Label']

print(f"Features shape (X): {X.shape}")
print(f"Target shape (Y):   {Y.shape}")
```

```python
## HERE WE HAVE TO APPLY S.M.O.T.E but only to TRAIN SPLIT

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'Y_train distribution: {Y_train.value_counts()}')
```

```python
#Now we have to encode using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print(f'X_train scaled shape: {X_train_scaled.shape}')
print(f"Mean of first feature (should be ~0): {X_train_scaled[:, 0].mean():.4f}")
print(f"Std of first feature (should be ~1):  {X_train_scaled[:, 0].std():.4f}")
```

```python
# SMOTE --> Synthetic Minority Oversampling Technique
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_smote, Y_train_smote = smote.fit_resample(X_train_scaled, Y_train)
print(f'BEFORE SMOTE SHAPE: {X_train_scaled.shape}')
print(f'AFTER SMOTE SHAPE: {X_train_smote.shape}')
print(f"\nClass distribution after SMOTE:")
print(pd.Series(Y_train_smote).value_counts())
```
